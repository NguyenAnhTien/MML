```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (30, 20)
plt.rcParams.update({'font.size': 40})
```

```python
x_data = []
y_data = []
for index in range(300):
    x_data.append(index)
    y = 2 * index + 100
    y_data.append(y)
```

```python
print(x_data[0], y_data[0])
```

```python
offset = (-10, 10)
plt.scatter([0, x_data[20], x_data[200], x_data[230]], [100, y_data[20], y_data[200], y_data[230]], s=800, c='green')
plt.plot(x_data, y_data, linewidth=10)
plt.annotate(f'A({x_data[0]}, {y_data[0]})', xy =(x_data[0], y_data[0]), xytext =(x_data[0] + offset[0], y_data[0] + offset[1]), fontsize=30)
plt.annotate(f'B({x_data[20]}, {y_data[20]})', xy =(x_data[20], y_data[20]), xytext =(x_data[20] + offset[0], y_data[20] + offset[1]), fontsize=30)
plt.annotate(f'C({x_data[200]}, {y_data[200]})', xy =(x_data[200], y_data[200]), xytext =(x_data[200] + offset[0], y_data[200] + offset[1]), fontsize=30)
plt.annotate(f'D({x_data[230]}, {y_data[230]})', xy =(x_data[230], y_data[230]), xytext =(x_data[230] + offset[0], y_data[230] + offset[1]), fontsize=30)
# plt.plot(0, 100, 'ro', linewidth=30)
```
