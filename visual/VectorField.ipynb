```python
import matplotlib.pyplot as plt
import numpy
```

```python
x_range = numpy.linspace(0, 2*numpy.pi, num=10)
y_range = numpy.linspace(0, 2*numpy.pi, num=10)
```

```python
x, y = numpy.meshgrid(x_range, y_range)
```

```python
def gradient(x, y):
    z_x = x * exp(-(x**2+y**2)/2) * (x**2 + y**3 - 3)
    z_y = y * exp(-(x**2+y**2)/2) * (x**2 + y**3 - 4)
    return (z_x, z_y)
```

```python
fx = x * numpy.exp(-(x**2+y**2)/2) * (x**2 + y**3 - 3)
```

```python
fy = y * numpy.exp(-(x**2+y**2)/2) * (x**2 + y**3 - 4)
```

```python
plt.quiver(x, y, fx, fy)
```
